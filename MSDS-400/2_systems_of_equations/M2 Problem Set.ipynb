```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
# Use sympy package to display the equation
sym.init_printing()
x,y = sym.symbols('x,y')

solns = sym.solve([
    -2*x + 1*y - 3,   # Set equations equal to 0 for solving
    3*x + 3*y + 9],
    [x, y])
print(f"Intersectiion is ({solns[x]}, {solns[y]}) ")


# Eq 1
# 2x + y = 1
# y = -2x + 1
# Eq 2
# 3x + 3y = - 3
# 3y = -3x - 3
# y = -x - 1

x_vals = np.arange(-5, 5)
y1 = 2 * x_vals + 3  # Eq 1
y2 = -x_vals - 3  # Eq 2

plt.plot(x_vals, y1)
plt.plot(x_vals, y2)
plt.show()
```

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
sym.init_printing()
x,y = sym.symbols('x,y')

solns = sym.solve([
    130*x - y,   
    50000 + 18*x - y],
    [x, y])
print(f"Intersection is ({solns[x]}, {solns[y]}) ")

x_vals = np.arange(0, 1000)
y1 = 130 * x_vals  # Eq 1
y2 = 50000 + 18*x_vals  # Eq 2

plt.plot(x_vals, y1)
plt.plot(x_vals, y2)
plt.title('Solving for Breakeven Point')
plt.legend(['R(x) = 130x', 'C(x) = 50000+18x'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import sympy as sym
a, b, c, d = sym.symbols('a, b, c, d')
eq1 = sym.Eq(a + b + c + d, 275)             
eq2 = sym.Eq(a - 4*b, 0) 
eq3 = sym.Eq(c - 5*d, 0)     
solns = sym.solve((eq1, eq2, eq3), (a, b, c, d))
print (solns)
```

<div class="alert alert-block alert-success"><b>Problem 9:</b></div>

```python
rhs = [119, 136099, 0]
rhs = np.matrix(rhs)
rhs = np.transpose(rhs)
print('\nRight Hand Side of Equation')
print(rhs)

A = [[1, 1, 1],
     [623, 1377, 1876],
     [1, -2, 0]]
A = np.matrix(A)
print('\nMatrix A')
print(A)
IA = np.linalg.inv(A)

result = np.dot(IA, rhs)
result = np.rint(result)
result
```

<div class="alert alert-block alert-success"><b>Problem 10:</b></div>

```python
blends = [[.65, .6, .95], 
          [.35, .4, .05]]
A = np.matrix(blends)
print('\nMatrix A')
print(A)


print('\nMatrix B')
demands = [[9, 9],
           [5, 2],
           [34, 34]]

B = np.matrix(demands)
print(B)

prod = np.dot(A, B)
print('\nProduct')
print(prod)

# Because these are in thousands of pounds
thous_prod = 1000 * prod

prices = [3.34, 0.56]
cost = np.dot(prices, thous_prod)
print("\nCost")
print(cost)
```

```python
lhs = np.matrix([[5, -5], [-1, 3]])
rhs = [3, -1]

print("Coefficient Matrix")
print(lhs)

print("Inverse of it is")
inv_lhs = np.linalg.inv(lhs)
print(inv_lhs)

sol = np.dot(inv_lhs, rhs)
print(f"The solutions are {sol}")
```
